```python
df = pd.read_excel('CASE 2024 - Seleção Data Analytics.xlsx', sheet_name='inf_municipais')
df
```

### Análise 1 - Agrupar cidades por ano e extrair a **média** das métricas numéricas elencadas para análise.

```python
cols = ['PESOURB', 'ESPVIDA', 'MORT1', 'E_ANOSESTUDO', 'T_ANALF25M', 'GINI', 'PIND', 'RDPC', 'IDHM']
df_group = df.groupby(['NOMEMUN', 'UF', 'REGIÃO', 'PORTE'], as_index=False)[cols].mean()
df_group
```

### Análise 2 - Eliminar colunas com métricas de qualidade correlatas.

Todas as colunas são bem correlacionadas exceto o índice GINI. Poderemos eliminar alguma dessas colunas pela redundância.

O GINI não estar correlacionado pode se dar pelo fato da diferença de escalas

```python
df_corr = df_group.loc[:, 'ESPVIDA':].corr()
sns.heatmap(df_corr, vmin=-1, vmax=1, center=0, annot=True, cmap='coolwarm_r');
```

### Análise 3 - Verificar ordem das melhores cidades usando-se das métricas de qualidade.

Eliminando as redundâncias, usaremos como métrica de qualidade: **IDHM, GINI e RDP**.

Ordenaremos também por **população urbana** pois há cidades muito pequenas com boa qualidade de vida porém dificilmente serão uma boa opção para instalação de um mercado.

Serão eliminadas também as cidades de porte **"Pequeno I" e "Pequeno II"**, depois será coletada as 100 melhores cidades.

```python
df_melhores_cidades = df_group.sort_values(by=['IDHM', 'RDPC', 'GINI', 'PESOURB'], ascending=[0, 0, 0, 0])
df_melhores_cidades = df_melhores_cidades[~df_melhores_cidades.PORTE.str.startswith('Pequeno')]
df_melhores_cidades = df_melhores_cidades.iloc[:100]
df_melhores_cidades.round(2).head(20)
```

```python
# Estados com maior potencial
df_melhores_cidades.UF.value_counts(ascending=True).plot(kind='barh');
```

### Análise 4 - Filtro pelos melhores estados de atuação

Verificando que os melhores estados são: **SP, SC, RS e MG**, vamos descartar cidades de outros estados da lista de potenciais cidades para implantação de um novo mercado.
